```python
from google.cloud import bigquery
```

```python
# Create a "Client" object
client = bigquery.Client()
```

```python
# Construct a reference to the "google_analytics_sample"
dataset_ref = client.dataset("google_analytics_sample", project="bigquery-public-data")

# API request - fetch the dataset
dataset = client.get_dataset(dataset_ref)
```

```python
# List  and print all the tables in the "google_analytics_sample" dataset
tables = list(client.list_tables(dataset))

for table in tables:  
    print(table.table_id)
```

```python
# Print information on all the columns in the "google_analytics_sample datasetusing for example ga_sessions_20160801
# Construct a reference to the "ga_sessions_20160801" table
table_ref = dataset_ref.table("ga_sessions_20160801")

# API request - fetch the table
table = client.get_table(table_ref)
table.schema
```

```python
# Preview the first five lines of the "full" table
client.list_rows(table, max_results=5).to_dataframe()
```
